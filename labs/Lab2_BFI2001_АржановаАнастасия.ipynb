```python
#Бинарный поиск
start_time = time.time()
def bin_index(mas, key):# находим индекс 
    n = len(mas)# записываем длину массива
    lower, upper = 0, n - 1# записываем минимальное и максимальное значения
    middle = (lower + upper) // 2 # записываем середину
    while upper - lower > 1: # пока разница между макс значением и мин больше 1
        if key <= mas[middle]: # проверяем что ключ который мы ввели меньше или равен заначению из списка под итндексом середина
            upper = middle# заменяем верхнюю границу на середину
        else:
            lower = middle# заменяем нижную границу на середину
        middle = (lower + upper) // 2#считаем середину
    return upper#возвращаем верхнее значение

def bin_find(mas, key):
    k = bin_index(mas, key)# записываем индекс где находится значение ключа
    return k
time_binary_search = round((time.time() - start_time)*1000)

def bin_insert(mas, key):#вставка
    mas.insert(bin_index(mas, key), key)#вставляем ключ на место индекса
    return mas

# Удаление
def bin_remove(mas, key):
    elem = bin_find(mas, key)
    mas.pop(elem)#вырезаем элемент под индексом элемент
    return mas

#print(bin_insert([5,13,19,26,30], 7))
chislo = randint(0,100)
print('Отсортированный исходный массив: ')
print(sorted_massiv)
t = str(chislo)+ ' '
print('Вставка элемента '+t+'в массив: ')
print(bin_insert(sorted_massiv, chislo))

print('Поиск элемента '+t+'в массиве:')
print(bin_find(sorted_massiv, chislo))

print('Удаление элемента '+t+'в массиве: ')
print(bin_remove(sorted_massiv, chislo))

print('Время работы поиска элемента: ', "--- {0} ms ---".format(time_binary_search))
```

```python
#Бинарное дерево
class node:
    def __init__(self,value=None):# задаем переменные в нон
        self.value=value
        self.left_child=None
        self.right_child=None
        self.parent=None # указатель на родительский узел в дереве

class binary_search_tree:# класс для бинарного дерева
    def __init__(self):
        self.root=None
#вставка
    def insert(self,value):
        if self.root==None:
            self.root=node(value)
        else:
            self._insert(value,self.root)

    def _insert(self,value,cur_node):
        if value<cur_node.value:
            if cur_node.left_child==None:
                cur_node.left_child=node(value)
                cur_node.left_child.parent=cur_node # set parent
            else:
                self._insert(value,cur_node.left_child)
        elif value>cur_node.value:
            if cur_node.right_child==None:
                cur_node.right_child=node(value)
                cur_node.right_child.parent=cur_node # set parent
            else:
                self._insert(value,cur_node.right_child)
        else:
            print("Value already in tree!")

    def print_tree(self):# вывод дерева
        if self.root!=None:#проверка что не пусто
            self._print_tree(self.root)

    def _print_tree(self,cur_node):
        if cur_node!=None:
            self._print_tree(cur_node.left_child)# вывод следующего левого значения
            print (str(cur_node.value))
            self._print_tree(cur_node.right_child)# вывод сдел правого значения

    def height(self):
        if self.root!=None:
            return self._height(self.root,0)
        else:
            return 0

    def _height(self,cur_node,cur_height):
        if cur_node==None: return cur_height
        left_height=self._height(cur_node.left_child,cur_height+1)
        right_height=self._height(cur_node.right_child,cur_height+1)
        return max(left_height,right_height)

    def find(self,value):
        if self.root!=None:
            return self._find(value,self.root)
        else:
            return None

    def _find(self,value,cur_node):
        if value==cur_node.value:
            return cur_node
        elif value<cur_node.value and cur_node.left_child!=None:
            return self._find(value,cur_node.left_child)
        elif value>cur_node.value and cur_node.right_child!=None:
            return self._find(value,cur_node.right_child)

#удаление
    def delete_value(self,value):
        return self.delete_node(self.find(value))

    def delete_node(self,node):

        ## -----
        # Improvements since prior lesson

        # Protect against deleting a node not found in the tree
        if node==None or self.find(node.value)==None:
            print("Node to be deleted not found in the tree!")
            return None 
        ## -----

        # returns the node with min value in tree rooted at input node
        def min_value_node(n):
            current=n
            while current.left_child!=None:
                current=current.left_child
            return current

        # returns the number of children for the specified node
        def num_children(n):
            num_children=0
            if n.left_child!=None: num_children+=1
            if n.right_child!=None: num_children+=1
            return num_children

        # get the parent of the node to be deleted
        node_parent=node.parent

        # get the number of children of the node to be deleted
        node_children=num_children(node)

        # break operation into different cases based on the
        # structure of the tree & node to be deleted

        # CASE 1 (node has no children)
        if node_children==0:

            # Added this if statement post-video, previously if you 
            # deleted the root node it would delete entire tree.
            if node_parent!=None:
                # remove reference to the node from the parent
                if node_parent.left_child==node:
                    node_parent.left_child=None
                else:
                    node_parent.right_child=None
            else:
                self.root=None

        # CASE 2 (node has a single child)
        if node_children==1:

            # get the single child node
            if node.left_child!=None:
                child=node.left_child
            else:
                child=node.right_child

            # Added this if statement post-video, previously if you 
            # deleted the root node it would delete entire tree.
            if node_parent!=None:
                # replace the node to be deleted with its child
                if node_parent.left_child==node:
                    node_parent.left_child=child
                else:
                    node_parent.right_child=child
            else:
                self.root=child

            # correct the parent pointer in node
            child.parent=node_parent

        # CASE 3 (node has two children)
        if node_children==2:

            # get the inorder successor of the deleted node
            successor=min_value_node(node.right_child)

            # copy the inorder successor's value to the node formerly
            # holding the value we wished to delete
            node.value=successor.value

            # delete the inorder successor now that it's value was
            # copied into the other node
            self.delete_node(successor)

            
#поиск
    start_time = time.time()
    def search(self,value):
        if self.root!=None:
            return self._search(value,self.root)
        else:
            return False

    def _search(self,value,cur_node):
        if value==cur_node.value:
            return True
        elif value<cur_node.value and cur_node.left_child!=None:
            return self._search(value,cur_node.left_child)
        elif value>cur_node.value and cur_node.right_child!=None:
            return self._search(value,cur_node.right_child)
        return False 
    
time_binary_search_tree = round((time.time() - start_time)*1000)# запись времени для выполнения функции
def fill_tree(tree, num_elems=10, max_int=100):# создание дерева
    from random import randint
    for _ in range(num_elems):
        cur_elem = randint(0, max_int)
        tree.insert(cur_elem)
    return tree
chislo1 = randint(0,100)
chislo2 = randint(0,100)

tree = binary_search_tree()#управшеная запись ссылки на класс
tree = fill_tree(tree) 

print ("Исходное дерево")
tree.print_tree()
print("результаты значения  после вставки "+str(chislo1)+":",tree.search(chislo1),", после вставки "+str(chislo2)+":",tree.search(chislo2))

print ("вставка элемента", chislo1)
tree.insert(chislo1)
tree.print_tree()
print ("результат значения после вставки "+str(chislo1)+":",tree.search(chislo1), ", после вставки "+str(chislo2)+":", tree.search(chislo2))
print ("удалеение значения", chislo1)
tree.delete_value(chislo1)
tree.print_tree()
print ("результат поиска посде удаления "+str(chislo1)+":", tree.search(chislo1), ", после удаления "+str(chislo2)+":",tree.search(chislo2))

print('Время работы поиска элемента: ',"--- {0} ms ---".format(time_binary_search_tree))
```

```python
#Метод Фибоначчи
class FibonacciSearch:
    def __init__(self):
        self.i = 0# задаем значения для класса
        self.p = 0
        self.q = 0
        self.stop = False

    def get_fibonacci_number(self, k):
        first_number = 0
        second_number = 1
        n = 0
        while n < k:
            temp_number = second_number
            second_number += first_number 
            first_number = temp_number
            n = n + 1
        return first_number

    def start_init(self, mas):
        self.stop = False
        k = 0
        n = len(mas)
        while (self.get_fibonacci_number(k+1) < n):
            k += 1
        m = self.get_fibonacci_number(k+1)-(n+1)
        self.i = self.get_fibonacci_number(k) - m
        self.p = self.get_fibonacci_number(k-1)
        self.q = self.get_fibonacci_number(k-2)

    def up_index(self):
        if self.p == 1:
            self.stop = True
        self.i = self.i + self.q
        self.p = self.p - self.q
        self.q = self.q - self.p

    def down_index(self):# переходим левее по ряду фибоначи
        if self.q == 0:
            self.stop = True
        self.i = self.i-self.q
        temp = self.q
        self.q = self.p - self.q
        self.p = temp   
        
    start_time = time.time()
#ПОИСК
    def search_index(self, mas, element):
        self.start_init(mas)
        while not self.stop:
            if self.i < 0:
                self.up_index()
            elif self.i >= len(mas):
                self.down_index()
            elif mas[self.i] == element:# если эмемент по индексу ай равен введеному искомому элементу
                result_index = self.i# результативный индекс равен ай и выходим у нас есть значение индекса нашего элемента
                break
            elif element < mas[self.i]:# если элемент меньше значения из массива
                self.down_index() # упускаем индекс, по ряду фибоначи переходим к более левым цифрам
            elif element > mas[self.i]:# если элемен больше
                self.up_index()#
                result_index = self.i +1#
        return result_index
    
    def find(self, mas, element):
        k = self.search_index(mas, element)
        assert element == mas[k], "Искомый элемент не найден!"
        return k    

    time_FibonacciSearch = round((time.time() - start_time)*1000)
#ВСТАВКА
    def insert(self, mas, element):# находим элемент куда мы вставляем значение и производим вставку
        mas.insert(self.search_index(mas, element), element)
        return mas             
    
#УДАЛЕНИЕ
    def remove(self, mas, element):
        el = self.find(mas, element)
        mas.pop(el)
        return mas

FS = FibonacciSearch()# объявление класса
chislo = randint(0,100)
print('Отсортированный массив',sorted_massiv)

print('Вставка элемента '+str(chislo)+' в массив',FS.insert(sorted_massiv, chislo))

print('Поиск элемента '+str(chislo)+' в массиве',FS.find(sorted_massiv, chislo))

print('Удаление элемента '+str(chislo)+' в массиве',FS.remove(sorted_massiv, chislo))

print('Время работы поиска элемента',"--- {0} ms ---".format(FS.time_FibonacciSearch))
```

```python
#Интерполяционный
start_time = time.time()

def interpolation_search_index(lys, val):
    low = 0
    high = (len(lys) - 1)
    while low <= high and val >= lys[low] and val <= lys[high]:
        index = low + int(((float(high - low) / ( lys[high] - lys[low])) * ( val - lys[low])))
        if lys[index] == val:
            return index
        if lys[index] < val:
            low = index + 1;
            guess = low
        else:
            high = index - 1;
            guess = high+1
    return guess

def interpolation_search_find(lst, search_item):
    k = interpolation_search_index(lst, search_item)
    assert search_item == lst[k], "Искомый элемент не найден"
    return k

time_interpolation_search = round((time.time() - start_time)*1000)

# Вставка
def interpolation_search_insert(lst, search_item):
    lst.insert(interpolation_search_index(lst, search_item), search_item)
    return lst

# Удаление
def interpolation_search_remove(lst, search_item):
    element = interpolation_search_find(lst, search_item)
    lst.pop(element)
    return lst

chislo = randint(0,100)
print('Отсортированный массив',sorted_massiv)

print('Вставка элемента '+str(chislo)+' в массив',interpolation_search_insert(sorted_massiv, chislo))

print('Поиск элемента '+str(chislo)+' в массиве',interpolation_search_find(sorted_massiv, chislo))

print('Удаление элемента '+str(chislo)+' в массиве',interpolation_search_remove(sorted_massiv, chislo))

print('Время работы поиска элемента',"--- {0} ms ---".format(FS.time_FibonacciSearch))
```

```python
#Метод цепочек
class ListNode:
    def __init__(self, key, val):
        self.pair = (key, val)
        self.next: Union[None, ListNode] = None
            
class ChainedHashTable:

    def __init__(self):
        self.m = 1000;
        self.h: list[Union[None, ListNode]] = [None]*self.m

    def __setitem__(self, key, value):
        index = self._hash(key)

        cur = self.h[index]
        if cur is None:
            self.h[index] = ListNode(key, value)

        else:
            while True:
                if cur.pair[0] == key:
                    cur.pair = (key, value)
                    return

                if cur.next == None: break
                cur = cur.next

            cur.next = ListNode(key, value)

    def __getitem__(self, key):
        index = self._hash(key)
        cur = self.h[index]

        while cur:
            if cur.pair[0] == key:
                return cur.pair[1]
            else:
                cur = cur.next
        raise KeyError(f"No such key: {key}")

    def _hash(self, key) -> int:
        return key % self.m
def test_hashing(hashable) -> None:
    tests_dict: dict[int, str] = dict({
        54: "cat",
        26: "dog",
        44: "lion",
        15: "tiger",
        20: "duck"
    })

    for k, v in tests_dict.items():
        hashable[k] = v

    for k, v in tests_dict.items():
        assert hashable[k] == v
print("Testing chained hash table")
test_hashing(ChainedHashTable())
print("Chained hash table tests passed")
```

```python
board = [[0 for i in range(8)] for j in range(8)]# создаем доску ввиде матрицы
count = 0#счетчик в нуле

def setQueen(i, j): # постановка королевы
    for x in range(8):
        board[x][j] += 1
        board[i][x] += 1
        if 0 <= i+j-x < 8:
            board[i+j-x][x] += 1
        if 0 <= i-j+x < 8:
            board[i-j+x][x] += 1
    board[i][j] = -1
    
def removeQueen(i, j):
    for x in range(8):
        board[x][j] -= 1
        board[i][x] -= 1
        if 0 <= i+j-x < 8:
            board[i+j-x][x] -= 1
        if 0 <= i-j+x < 8:
            board[i-j+x][x] -= 1
    board[i][j] = 0
    
def printPosition():
    global count
    count += 1
    abc = 'abcdefgh'
    ans = []
    for i in range(8):
        for j in range(8):
            if board[i][j] == -1:
                ans.append(abc[j]+str(i+1))
    print(', '.join(ans))
    
def solve(i):#
    for j in range(8):#
        if board[i][j] == 0:# если на этом месте пусто
            setQueen(i, j)#переходим в поставление королевы
            if i == 7:# если i == 7 производим вывод
                printPosition()
            else:
                solve(i+1)#увеличиваем значение i на 1 и сново заходим в ту же ф-ю
            removeQueen(i, j)#
            
print("Варианты расстановок")
solve(0)
print("Найденное количество расстановок",count)
```
