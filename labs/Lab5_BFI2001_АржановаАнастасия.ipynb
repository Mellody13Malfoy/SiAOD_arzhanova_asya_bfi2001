```python
#Кривая Коха
import turtle

def drow_koch_seg(t,ln):
    if ln>6:
        ln_new = ln//3
        drow_koch_seg(t,ln_new)
        t.left(60)
        drow_koch_seg(t,ln_new)
        t.right(120)
        drow_koch_seg(t,ln_new)
        t.left(60)
        drow_koch_seg(t,ln_new)        
    else:
        t.forward(ln)
        t.left(60)
        t.forward(ln)
        t.right(120)
        t.forward(ln)
        t.left(60)
        t.fd(ln)        

width = 1200
heidht = 800
screen = turtle.Screen()
screen.setup(width, heidht,0,0)

t = turtle.Turtle()
t.ht()#скрываем черепашку
t.speed(100)# ускорение черепашки
drow_koch_seg(t,150)
turtle.done()
```

```python
#Кривая Гильберта
# -*- coding: utf-8 -*-
import turtle           

def A(size, n):
    if n==0:          # A не рисуем никак
        return
    t.left(90)        # -
    B(size, n-1)      # B
    t.fd(size)        # F
    t.right(90)       # +
    A(size, n-1)      # A
    t.fd(size)        # F
    A(size, n-1)      # A
    t.right(90)       # +
    t.fd(size)        # F
    B(size, n-1)      # B
    t.left(90)        # -

def B(size, n):
    if n==0:          # B не рисуем никак
        return
    t.right(90)       # +
    A(size, n-1)      # A 
    t.fd(size)        # F
    t.left(90)        # -
    B(size, n-1)      # B
    t.fd(size)        # F
    B(size, n-1)      # B 
    t.left(90)        # -
    t.fd(size)        # F
    A(size, n-1)       # A    
    t.right(90)       # +
    
t = turtle.Turtle()
t.shape("turtle")
t.ht()#скрываем черепашку
t.width(3)
t.speed(100)

A(40, 3)     # правила применили глубиной 10

t.hideturtle()
turtle.done()
```

```python
#кривая Серпинского
import turtle


def serp_out (size, n):
    serp(size, n, "end")
    t.right(90 + 45)
    t.forward(size/pow(2, n))

def serp (size, n, pos):
    if n == 0:
        t.left(45)
        t.fd(size)
        t.right(90)
        t.fd(size)
        t.right(90)
        t.fd(size)
        t.left(45)
    else:
        if pos == 'diag up': 
            if n != 1:
                serp(size/2, n-1, "diag up")
            
            t.left(45)
            t.fd(size/pow(2, n))
            t.left(45)
            t.fd(size/pow(2, n-1))
            serp(size, n, "end")
            t.fd(size/pow(2, n-1))
            if n != 1:
                serp(size/2, n-1, "diag up")
        elif pos == 'diag down':
            #right up
            if n != 1:
                serp(size/2, n-1, "diag down")
            #right down
            t.forward(size/pow(2, n-1))
            serp(size, n, "end")
            t.forward(size/pow(2, n-1))
            t.left(45)
            t.forward(size/pow(2, n))
            t.left(45)
            #left down
            if n != 1:
                serp(size/2, n-1, "diag down")
            
        elif pos == 'center':
            #left up
            t.left(45)
            t.forward(size/pow(2, n))
            t.left(45)
            t.forward(size/pow(2, n-1))
            serp(size, n, "end")
            #right up
            t.forward(size/pow(2, n-1))
            serp(size, n, "end")
            #left down
            t.forward(size/pow(2, n-1))
            serp(size, n, "end")
            #right down
            t.forward(size/pow(2, n-1))
            t.left(45)
            t.forward(size/pow(2, n))
            t.left(45)
            
        else:
            if n > 2:
                serp(size/4, n-2, "diag up")
            #else
            serp(size/2, n-1, "center")
            #left down
            if n > 2:
                serp(size/4, n-2, "diag down")
                
size = 100
n = 4
t = turtle.Turtle()
t.clear()
t.setheading(0)
t.speed(10)

serp_out(size, n)
turtle.mainloop()
```

```python
#Салфетка Серпинского
import turtle
def A(size, n):
    if n==0:          # A не рисуем никак
        return
    B(size, n-1)      # B
    t.fd(size)        # F
    t.left(60)       # +
    A(size, n-1)      # A
    t.fd(size)        # F
    t.left(60)       # +
    B(size, n-1)      # B

def B(size, n):
    if n==0:          # B не рисуем никак
        return
    A(size, n-1)      # A 
    t.fd(size)        # F
    t.right(60)        # -
    B(size, n-1)      # B
    t.fd(size)        # F
    t.right(60)        # -
    A(size, n-1)       # A
    
t = turtle.Turtle()
t.shape("turtle")
t.ht()#скрываем черепашку
t.width(3)
t.speed(100)

A(10, 5)     # правила применили глубиной 10

t.hideturtle()
turtle.done()
```

```python
#Ковер Серпинского
import turtle

def s(l,n):
    if n == 0:
        t.color('black')#объявляем цвет
        t.begin_fill()#вызывается непосредственно перед заливкой,
        for i in range (4):#рисуем квадрат
            t.fd(l)
            t.left(90)
        t.end_fill()#заливает фигуру цветом

    else:
        for i in range(4):
            #первая часть большого кводрата
            s(l/3,n-1)    
            t.fd(l/3)
            #серединые кводраты
            s(l/3,n-1)     
            t.fd(l/3)
            # оставшееся часть
            t.fd(l/3)
            t.left(90)

        turtle.update()
t = turtle.Turtle()
t.ht()#скрываем черепашку
t.speed(100)# ускорение черепашки

s(400, 3)
turtle.done()
```

```python
#Дерево Пифагора
import turtle
import math
def pifagor_tree(size, depth):
    if depth == 0:
        t.left(90)
        for i in range(0,3):
            t.fd(size*math.sqrt(2)/2)
            t.right(90)
        t.fd(size*math.sqrt(2)/2)
    else:
        t.left(90)
        t.fd(size)
        t.right(45)
        if depth == 1:
            pifagor_tree(size, depth-1)
        else:
            pifagor_tree(size*math.sqrt(2)/2, depth-1)
        t.right(90)
        t.right(135)
        t.fd(size)
        t.left(45)
        if depth == 1:            
            pifagor_tree(size, depth-1)
            t.right(90)
            t.right(225)
        else:
            t.left(90)
            t.fd(size*math.sqrt(2)/2)
            t.right(180)
            pifagor_tree(size*math.sqrt(2)/2, depth-1)
            t.right(180)
            t.forward(size*math.sqrt(2)/2)
            t.right(45)
        t.fd(size)
        t.right(90)
        t.fd(size)
def half_sierpinski(size, level):
    if level == 0: 
        t.fd(size)
        return
    half_sierpinski(size, level - 1)
    t.left(45)
    t.fd(size*math.sqrt(2)) 
    t.left(45)
    half_sierpinski(size, level - 1)
    t.right(90)
    t.fd(size)
    t.right(90)
    half_sierpinski(size, level - 1)
    t.left(45)
    t.fd(size*math.sqrt(2)) 
    t.left(45)
    half_sierpinski(size, level - 1)

def sierpinski_line(size, level):
    half_sierpinski(size, level)
    t.right(90)
    t.fd(size)
    t.right(90)
    half_sierpinski(size, level)
    t.right(90)
    t.fd(size)
    t.right(90)
t = turtle.Turtle()
t.shape("turtle")
t.ht()#скрываем черепашку
t.width(3)
t.speed(100)
pifagor_tree(100,3)
t.hideturtle()
turtle.done()
```
