```python
def read():
    f = open("10.txt", "r")
    s=""
    record = []
    for i in f:
        for ii in i:
            if ii != "\n":
                s += ii
        record.append(s.split())
        s = ""
    f.close()
    t = int(record[0][0])
    n = []
    a = []
    b = []
    for i in range(0,t):
        n.append(int(record[1+i][0]))
        a.append(int(record[1+i][1]))
        b.append(int(record[1+i][2]))
    return t,n,a,b

def record(a):
    f = open("answer_10.txt","w")
    for i in a:
        f.write(str(i)+"\n")
    f.close

def enumeration(k,a,b,n,q):
    if n<k:
        q.append('NO')
        return
    if (n-k)%b==0: 
        q.append('YES')
        return
    else: 
        enumeration(k*a,a,b,n,q)  
        

le,n,a,b = read()
answer = []
for i in range(le):
    q=[]
    flag = ''
    if a[i] != 1:
        enumeration(1,a[i],b[i],n[i],q)
        flag = " ".join(q)
    else:
        flag = "YES"
    answer.append(flag)
    print(flag)
record(answer)   
    
```
