### Задача 4
Дано целое неотрицательное число n, запись которого состоит из не более, чем 100 цифр, и не содержит ведущих незначащих нулей.

Нужно определить, можно ли в данном числе вычеркнуть некоторое (возможно нулевое) количество цифр так, чтобы полученное после вычеркивания цифр число содержало хотя бы одну цифру, было неотрицательным, не имело ведущих незначащих нулей и делилось на 8. После вычеркивания переставлять цифры запрещается.

Если решение существует, необходимо вывести его.

Входные данные\
В единственной строке входных данных содержится целое неотрицательное число n. Запись числа n не содержит ведущих незначащих нулей и ее длина не превосходит 100 цифр.

Выходные данные\
Выведите "NO" (без кавычек), если искомого способа вычеркнуть некоторые цифры из числа n не существует.

Иначе выведите "YES" в первой строке и число, получившееся в результате вычеркивания некоторых цифр числа n, во второй строке. Выведенное число должно делиться на 8.

Если возможных ответов несколько, требуется вывести любой из них.

Примеры\
входные данные\
3454\
выходные данные\
YES\
344\
входные данные\
10\
выходные данные\
YES\
0\
входные данные\
111111\
выходные данные\
NO

```python
def delit(n,w):
    if len(str(n)) == 1:
        w.append(-1)
        return
    elif n % 8 == 0:
        w.append(n)
        return
    else:
        for i in range(len(str(n))):
            t = int(str(n)[:i]+str(n)[i+1:])
            if  t% 8 == 0:
                w.append(t)
                return

            else:
                delit(t,w) 

def record(a):
    f = open("answer_4.txt","w")
    f.write(a+"\n")
    f.close

n_st = input("Введите число ")
flag = True
while flag:
    flag = False
    if not n_st.isdigit():
        n_st = input("Введите число ")
        flag = True
    elif len(n_st) > 100:
        n_st = input("Введите число длиной меньше 100 цифр")
        flag = True
n = int(n_st)
w = []
delit(n,w)
flag = False
answer  = 0 
w.sort()
w.reverse()
for i in w:
    if i != -1:
        answer = str(i)
        flag = True
        break
if flag:
    print("YES\n"+answer)
    record("YES\n"+answer)
else:
    print("NO")
    record("NO")
```

Есть n деревьев, расположенных вдоль дороги в точках с координатами x1, x2, ... , xn. Каждое дерево имеет свою высоту hi. Дерево можно срубить и повалить влево или вправо. Дерево можно повалить, если на отрезке, который оно должно занимать после сваливания, нет ни одной занятой точки. В последующих n строках находятся пары целых чисел xi, hi (1 ≤ xi, hi ≤ 109) — координата и высота і-го дерева.

Пары заданы в порядке возрастания xi. Никакие два дерева не находятся в точке с одинаковой координатой.

Выходные данные\

В первом примере деревья можно повалить так:

повалить 1-е дерево влево — теперь оно занимает отрезок [ - 1;1]\
повалить 2-е дерево вправо — теперь оно занимает отрезок [2;3]\
оставить 3-е дерево — оно занимает точку 5\
оставить 4-е дерево — оно занимает точку 10\
повалить 5-е дерево вправо — теперь оно занимает отрезок [19;20]\
Во втором примере можно повалить еще и 4 дерево вправо, после чего оно займет отрезок [10;19].\

```python
def read():
    f = open("5.txt", "r")
    record=[]
    s=""
    for i in f:
        for ii in i:
            if ii != "\n":
                s += ii
        record.append(s)
        s = ""
    f.close()
    if not record[0].isdigit():
        return 0,0,True,"В файле в первой строке одно из знечений не число"    
    n = int(record[0])
    if n > 10**5:
        return 0,0,True,"В файле в первой строке одно из чисел больше 100000"
    elif n <1:
        return 0,0,True,"В файле в первой строке первое число меньше 1"
    x= []
    for i in range(n):
        t =record[1+i].split()
        t_1 = int(t[0])
        t_2 = int(t[1])
        if t_1 < 1 or t_2 <1:
            return 0,0,True,"Одно из значений меньше чем нужно. Введите числа больше 1"
        elif t_1 > 10**9 or t_2> 10**9:
            return 0,0,True, "Одно из значений больше чем нужно. Введите числа больше 10^5"
        x.append([t_1,t_2])
    return n,x,False,""

def record(a):
    f = open("answer_5.txt","w")
    f.write(str(a)+"\n")
    f.close

n,x,flag,mistake = read()
ch=0
if flag:
    print("Проверьте правильность введенных данных в файле.\nОшика:" + mistake)
else:
    
    for i in range(len(x)):
        if i == 0 or i == len(x)-1:
            ch+=1
        else:
            if x[i][0]+x[i][1]<x[i+1][0]:
                ch+=1
            elif x[i][0]-x[i][1]>x[i-1][0]:
                ch+=1
    print(ch)
    record(str(ch))
```

```python
def read():
    f = open("10.txt", "r")
    s=""
    record = []
    for i in f:
        for ii in i:
            if ii != "\n":
                s += ii
        record.append(s.split())
        s = ""
    f.close()
    if not record[0][0].isdigit() or not record[0][0].isdigit():
        return 0,0,0,0,True,"В файле в первой строке одно из знечений не число"         
    t = int(record[0][0])
    if t> 10**5:
        return 0,0,0,0,True,"В файле в первой строке одно из чисел больше 100000"
    elif t <1:
        return 0,0,0,0,True,"В файле в первой строке первое число меньше 1"
        
    n = []
    a = []
    b = []
    for i in range(0,t):
        t_1=int(record[1+i][0])
        t_2=int(record[1+i][1])
        t_3=int(record[1+i][2])
        if t_1 < 1 or t_2 <1 or t_3 <1 :
            return 0,0,True,"Одно из значений меньше чем нужно. Введите числа больше 1"
        elif t_1 > 10**9 or t_2> 10**9 or t_3> 10**9:
            return 0,0,True, "Одно из значений больше чем нужно. Введите числа больше 10^5"
                
        n.append(t_1)
        a.append(t_2)
        b.append(t_3)
    return t,n,a,b,False,""

def record(a):
    f = open("answer_10.txt","w")
    for i in a:
        f.write(str(i)+"\n")
    f.close

def enumeration(k,a,b,n,q):
    if n<k:
        q.append('NO')
        return
    if (n-k)%b==0: 
        q.append('YES')
        return
    else: 
        enumeration(k*a,a,b,n,q)  
        

le,n,a,b,flag,mistake = read()
answer = []
if flag:
    print("Проверьте правильность введенных данных в файле.\nОшика:" + mistake)
else:
    for i in range(le):
        q=[]
        flag = ''
        if a[i] != 1:
            enumeration(1,a[i],b[i],n[i],q)
            flag = " ".join(q)
        else:
            flag = "YES"
        answer.append(flag)
        print(flag)
    record(answer)   
    
```
